```python
def get_artifact_group(arts, index, n=1):
    """
    Returns up to 3 artifacts centered around the given index +/- n,
    adjusting to stay within bounds of the arts list.
    """
    try:
        if n >= 0:
            indices = [index + i for i in range(-n, n+1)]
        else:
            indices = [index + i for i in range(n, -n+1)]

        # Filter out-of-bounds indices
        valid_indices = [i for i in indices if 0 <= i < len(arts)]

        # Format artifacts
        link = "\n\n".join(
            f"ARTIFACT {i+1}: {arts[idx]}" for i, idx in enumerate(valid_indices)
        )

    except Exception as e:
        link = f"Error retrieving artifacts: {str(e)}"

    return link
```

```python
def arts_to_str(arts):
    s = ""
    i = 1
    for art in arts:
        s = s + f"ARTIFACT {i}: {art}\n\n"
        i += 1
    
    return s
```
