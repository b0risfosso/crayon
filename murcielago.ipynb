```python
from search import search_library
DB_FILE = "library_temp.sqlite"
MAX_RESULTS = 100
```

```python
# can you seed a person? 
KEYWORDS = ["person", "people"]  # list of words
search_results = search_library(DB_FILE, KEYWORDS, MAX_RESULTS, match_all=True)
# papers = {}
for r in search_results:
    title, text = r['title'], r['text']
    print(title)
```

```python
# can you seed dressmaking? 
KEYWORDS = ['"sewing"', '"needlework"', '"garment construction"', '"fashion design"', '"making dresses"', '"how to sew"']
search_results = search_library(DB_FILE, KEYWORDS, MAX_RESULTS, match_all=False)
# papers = {}
for r in search_results:
    title, text = r['title'], r['text']
    print(title)
```
