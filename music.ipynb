```python
from search import search_library
import random
from dotenv import load_dotenv
from openai import OpenAI, OpenAIError
import os

def ask(sys_msg, usr_msg):
    resp = client.chat.completions.create(
        model=OPENAI_MODEL,
        messages=[
            {"role": "system", "content": sys_msg},
            {"role": "user", "content": usr_msg}
        ]
    )
    return resp.choices[0].message.content

def sample_text(text, x):
    max_start = max(0, len(text) - x)
    random_start = random.randint(0, max_start)
    excerpt = text[random_start:random_start + x]
    return excerpt


load_dotenv()
client = OpenAI()
OPENAI_MODEL = os.getenv("OPENAI_MODEL", "o4-mini-2025-04-16")
DB_FILE = "library.sqlite"
MAX_RESULTS = 100
```

```python
KEYWORDS = ["hindgut morphogenesis", "hindgut", "active tension network"]
search_results = search_library(DB_FILE, KEYWORDS, MAX_RESULTS, match_all=False)
hindgut = {}
for r in search_results:
    title, text = r['title'], r['text']
    hindgut[title] = text


rem = ["Electrical Stimulation for Pelvic Floor Disorders", None, "Fundamental Neuroscience (3rd edition)", "Clinical Emergency Medicine", "Harrison's Principles of Internal Medicine 15th Edition"]
for r in rem:
    if r in hindgut:
        del hindgut[r]

for title, text in hindgut.items():
    print(title)
```

```python
for _ in range(3):
    n = random.randint(0, len(hindgut)-1)
    title, text = list(hindgut.keys())[n], list(hindgut.values())[n]
    excerpt = sample_text(text, 10000)
    act1 = ask("If I have a lightsheet imaging dataset of hindgut development over time, what hypothesis driven inquires should be performed based the following information?", excerpt)
    act2 = ask("If I have a 2D movie of the developing Drosophila hindgut acquired by identifying and unraveling the apical surface of the epithelial tissue of a lightsheet dataset of developing hindgut, what hypothesis driven inquires should be performed based on the following information?", excerpt)
    
    print("\n\nACT 1:\n", act1)
    print("\n\nACT 2:\n", act2)
```
