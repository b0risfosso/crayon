```python
from search import search_library
import random
from dotenv import load_dotenv
from openai import OpenAI, OpenAIError
import os

def ask(sys_msg, usr_msg):
    resp = client.chat.completions.create(
        model=OPENAI_MODEL,
        messages=[
            {"role": "system", "content": sys_msg},
            {"role": "user", "content": usr_msg}
        ]
    )
    return resp.choices[0].message.content

def sample_text(text, x):
    max_start = max(0, len(text) - x)
    random_start = random.randint(0, max_start)
    excerpt = text[random_start:random_start + x]
    return excerpt


load_dotenv()
client = OpenAI()
OPENAI_MODEL = os.getenv("OPENAI_MODEL", "o4-mini-2025-04-16")
DB_FILE = "library.sqlite"
MAX_RESULTS = 100
```

```python
KEYWORDS = ["missile"]
search_results = search_library(DB_FILE, KEYWORDS, MAX_RESULTS, match_all=False)
information = {}
for r in search_results:
    title, text = r['title'], r['text']
    information[title] = text


#rem = ["Electrical Stimulation for Pelvic Floor Disorders", None, "Fundamental Neuroscience (3rd edition)", "Clinical Emergency Medicine", "Harrison's Principles of Internal Medicine 15th Edition"]
#for r in rem:
#    if r in hindgut:
#        del hindgut[r]

for title, text in information.items():
    print(title)
```

```python
for _ in range(1):
    n = random.randint(0, len(information)-1)
    title, text = list(information.keys())[n], list(information.values())[n]
    excerpt = sample_text(text, 10000)
    summary = ask("Provide a one sentence summary of the following excerpt.", excerpt)
    dirt = ask("What hypothesis driven inquires can be performed based the following information?", excerpt)
    narrative = ask("Given the following set of hypotheses, describe the narrative for each hypotheses. Describe possible outcomes, what data confirms/falsified the hypothesis and the implications for future hypotheses.", dirt)
    
    print("\n\nDIRT:\n", summary, dirt, narrative)
```
